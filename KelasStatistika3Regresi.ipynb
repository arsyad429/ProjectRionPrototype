# library matplotlib



```python
import matplotlib.pyplot as plt
from scipy import stats
x = [5,7,8,7,2,17,2,9,4,11,12,9,6]#buat data variabel indepndent
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]# buat data variabel dependent



slope, intercept, r, p, std_err = stats.linregress(x, y)#regresi linier


#slope = gradien
#intercept = konstanta
#r = nilai kolerasi
#std_err =
#p =
print(slope, intercept, r, p, std_err)

print("y = ",slope,"x +", intercept)#persaman regresi
```

```python
x= [3,5,7,10,15]#independent
y = [40,36,33,28,22]#dependent

slope, intercept, r, p, std_err = stats.linregress(x, y)#regresi linier


#slope = gradien
#intercept = konstanta
#r = nilai kolerasi
#std_err =
#p =
print(slope, intercept, r, p, std_err)

print("y = ",slope,"x +", intercept)#persaman regresi

def myfunc(x):
  return slope * x + intercept

mymodel = list(map(myfunc, x))

plt.scatter(x , y)
plt.plot(x, mymodel, 'red')
plt.show()
```

# Library sklearn


```python
import numpy as np#supaya bisa pakai array
from sklearn.linear_model import LinearRegression
#panggil function array dengan bantuan library numpy
x = np.array([5, 15, 25, 35, 45, 55]).reshape((-1,1)) #reshape(-1,1) agar array menjadi baris/di transfus -> harus
y = np.array([5,20,14,32,22,38])

#LinearRegression().fit(x,y)#panggil fungsi LinearRegression->penulisan lain
model = LinearRegression()
model.fit(x,y)

print('intercept', model.intercept_)
print('slope', model.coef_)
print('y=',model.intercept_,'x',model.coef_)
r_sq = model.score(x, y) #koefisien determinasi
print('koefisien determinasi ', r_sq)

#gambar

plt.scatter(x, y, color='red')#plotting data
plt.plot(x, model.predict(x), color='blue')#ploting garis regresi
plt.title("gaji vs pengalaman (tahun)")#judul

plt.xlabel("pengalaman(tahun)")# label sb x
plt.ylabel("gaji") #label sb y

plt.show



```

# Library Statmodels

```python
import statsmodels.api as sm
y = [5,6,8,10,11]
x = [2,4,5,7,8]
#tranfus x
x = sm.add_constant(x)#add_constant untuk transfus matrix x
model = sm.OLS(y,x)#function untuk regresi linier(ordinary list square)
result = model.fit()#.fit untuk menghitung titik modelnya dimasukan ke variabel result
print(result.summary())#summary untuk menunjukan nilai apa saja yang bisa dijalankan stats model saat menggunakan OLS

```

```python
#gambar grafik
sm.graphics.plot_partregress_grid(result)
```
